```python
import matplotlib.pyplot as plt

# 시도별 의료기관 수 계산
hospital_count = df['시군명'].value_counts()
hospital_count
```

```python

# 시각화
plt.figure(figsize=(10, 6))
hospital_count.plot(kind='bar')
plt.xlabel('시도')
plt.ylabel('의료기관 수')
plt.title('시도별 의료기관 수')
plt.xticks(rotation=45)
plt.show()
```

# 쓰레기 자료

```python
# 깊은 복사를 통해 원본 df를 훼손하지 않고 df_1 생성 및 NaN 값 0으로 전처리
df_1 = df.copy()
df_1['응급의료지원센터여부'] = df_1['응급의료지원센터여부'].fillna('0')
df_1['전문응급의료센터여부'] = df_1['전문응급의료센터여부'].fillna('0')
df_1['전문응급센터전문분야'] = df_1['전문응급센터전문분야'].fillna('0')
df_1['권역외상센터여부'] = df_1['권역외상센터여부'].fillna('0')
df_1['지역외상센터여부'] = df_1['지역외상센터여부'].fillna('0')
df_1['소재지도로명주소'] = df_1['소재지도로명주소'].fillna('0')
df_1['소재지지번주소'] = df_1['소재지지번주소'].fillna('0')
df_1['소재지우편번호'] = df_1['소재지우편번호'].fillna(0)
df_1['위도'] = df_1['위도'].fillna(0)
df_1['경도'] = df_1['경도'].fillna(0)
df_1
```
